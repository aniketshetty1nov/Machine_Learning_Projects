```python
# In this below function I will remove the patterns in the input text
def removingPatterns(txtTweets,pttrn):
    myregexp = re.findall(pttrn,txtTweets)
    for word in myregexp:
        txtTweets = re.sub(word, "", txtTweets)
    return txtTweets
```

```python
#Now i will create function to  remove twitter handles i.e (@username etc)
mytwitterSentAnalysis['cleaned_tweets'] = np.vectorize(removingPatterns)(mytwitterSentAnalysis['tweets'], "@[\w]*")
```

```python
mytwitterSentAnalysis['cleaned_tweets'].head()
```

```python
#here i am replacing  all the sepcial characters, numbers and punctuations which are not alphabets
mytwitterSentAnalysis['cleaned_tweets'] = mytwitterSentAnalysis['cleaned_tweets'].str.replace("[^a-zA-Z#]"," ")
mytwitterSentAnalysis['cleaned_tweets'].head()
```

```python
# Lets remove now Stop words from the Tweets
mytwitterSentAnalysis['cleaned_tweets'] = mytwitterSentAnalysis['cleaned_tweets'].apply(lambda x: " ".join([w for w in x.split() if len(w)>3]))
mytwitterSentAnalysis['cleaned_tweets'].head()
```

```python
def wordopt(text):
    # Remove URls 
    text = re.sub(r'https?://\S+|www\.\S+', '', text)
#     #replace all non-alphanumeric characters in the text with space characters
#     text = re.sub("\\W"," ",text) 
    return text    

mytwitterSentAnalysis['cleaned_tweets'] = mytwitterSentAnalysis['cleaned_tweets'].apply(wordopt)
mytwitterSentAnalysis['cleaned_tweets']
mytwitterSentAnalysis.head(10)
```

```python
# Lets now split the tweets into list of words i.e tokens
myTweetsTokenized = mytwitterSentAnalysis['cleaned_tweets'].apply(lambda x: x.split())
myTweetsTokenized.head()
```

```python
# stem the words
from nltk.stem.porter import PorterStemmer
stemmer = PorterStemmer()

myTweetsTokenized = myTweetsTokenized.apply(lambda sentence: [stemmer.stem(word) for word in sentence])
myTweetsTokenized.head()
```

```python
# Combine to single sentence 
for i in range(len(myTweetsTokenized)):
    myTweetsTokenized[i] = " ".join(myTweetsTokenized[i])
    
mytwitterSentAnalysis['cleaned_tweets'] = myTweetsTokenized
mytwitterSentAnalysis.head()
```

```python
#Let now extract the hastag
def extractHashtag(tweets): 
    hashtags = [] # assign the List to an hashtag
    #loop words in the tweet
    for tweet in tweets:
        ht =re.findall(r"#(\w+)", tweet)
        hashtags.append(ht)
    return hashtags 
        
```

```python
# extract hastags from non racist/sexist tweets
positiveHashTags = extractHashtag(mytwitterSentAnalysis['cleaned_tweets'][mytwitterSentAnalysis['target']==1])

# extract hastags from racists/sexist Tweets
negativeHashTags = extractHashtag(mytwitterSentAnalysis['cleaned_tweets'][mytwitterSentAnalysis['target']==0])
```

```python
positiveHashTags
negativeHashTags
```

```python
# unnest List - lets now combine all the List into a Single list
positiveHashTags = sum(positiveHashTags,[])
negativeHashTags = sum(negativeHashTags,[])
```

```python
# lets use a dict from nktk package to process the tweets 
thefreqTwpos = nltk.FreqDist(positiveHashTags)
# then lets convert the above intoa. dataframe.
thefreqTwposdf = pd.DataFrame({'Hashtag': list(thefreqTwpos.keys()),
                  'Count' : list(thefreqTwpos.values())
                 })
thefreqTwposdf.head()
```

```python
# Splitting the data into training and testing data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, stratify=Y, random_state=2)
print(X.shape,X_train.shape, X_test.shape)
```

```python
print(X_test)
```

```python
print(X_train)
```

```python
#Feature Extraction
# Now I will use the Vectorization technique to convert the tweets into numeric form
myvectorizer = TfidfVectorizer()
X_train = myvectorizer.fit_transform(X_train)
X_test = myvectorizer.transform(X_test)
```

# Training Our Model

### 1) Logistic Model using SKLEARN

```python
# Logistic Regression using sklearn package.
skLrnLogisticModel = LogisticRegression(max_iter=1000)
skLrnLogisticModel.fit(X_train,Y_train)
```

```python
# Model Evaluation
#Let find out our accuracy
x_train_prediction =  skLrnLogisticModel.predict(X_train)
training_data_accuracy = accuracy_score(Y_train,x_train_prediction)
```

```python
print(f'The accuracy score on our training data of our logistic using skLearn model is: {training_data_accuracy} ')
```

```python
# Model Evaluation
#Let find out our accuracy
x_test_prediction =  skLrnLogisticModel.predict(X_test)
test_data_accuracy = accuracy_score(Y_test,x_test_prediction)
```

```python
# Y_test.shape
x_test_prediction.shape
```

```python
x_test_prediction
```

```python
print(f'The accuracy score on our test data of our logistic using skLearn model is: {test_data_accuracy} ')
print(f'The F1 score on our test data of our logistic using skLearn model is: {f1_score(Y_test, x_test_prediction)} ')
```

Model Accuracy - 77.8%

```python
# Compute confusion matrix
def PlotConusionMatrix(y_test, pred):
    cnfsnMatrx = confusion_matrix(y_test, pred)

    # Lets plot the confusion matrix 
    plt.figure(figsize=(5, 4))
    sns.heatmap(cnfsnMatrx, annot=True, fmt='d', cmap='gist_ncar_r', cbar=False,annot_kws={"fontsize": 14})
    plt.xlabel('Predicted labels',fontsize=14)
    plt.ylabel('True labels',fontsize=14)
    plt.title('Confusion Matrix for SKlearn Logistic Regression',fontsize=16)

    plt.show()

PlotConusionMatrix(Y_test,x_test_prediction) 

# The top left area displays the no. of truly negatives. i.e. values that were correctly predicted as negative by the model.
# The top-right area displays the no. of false positives i.e. incorrectly predicted as positive by the model
# The bottom-left area displays the no. of false negatives i.e. data which are incorrectly predicted as negative by the model.
# The bottom-right area displays the no. of truly positives i.e. correctly predicted as positive by the model.


# Here we can clearly observe that the Model correctly predicted the 126554(positve True) positive tweets but missed 
# 33446 of them.

# Here we can clearly observe that the Model correctly predicted the 126554(positve True) positive tweets but missed 
# 33446 of them.
# And on the other hand 121084 negative tweets were correctly guessed by the model but it failed 38916 to predict
```

```python
print(classification_report(Y_test, x_test_prediction))
```

```python
# Defining New Dict variables to store all out accuracy & F1 scores.
Accuracies = {}
F1Scores = {}

Accuracies['sklearn_Logis']= accuracy_score(Y_test, x_test_prediction)
F1Scores['sklearn_Logis']= f1_score(Y_test, x_test_prediction)
```
